```python
import plotly.express as px
import pandas as pd

# Load the data
file_path = 'Second_Step_BERTopic_topic_info_labelled_with_Final_Label_Merged.csv'
data = pd.read_csv(file_path)

# Prepare data for sunburst chart
sunburst_data_with_count = data[['Final_Label', 'Higher_Topic_Label', 'Human_Readable_Topic', 'Count']].copy()

# Calculate the total count for percentages
total_count = sunburst_data_with_count['Count'].sum()

# Calculate the percentage for each topic
sunburst_data_with_count['percentage'] = (sunburst_data_with_count['Count'] / total_count) * 100

# Add percentage to the labels for display in the chart
sunburst_data_with_count['Human_Readable_Topic'] = sunburst_data_with_count['Human_Readable_Topic'] + ' (' + sunburst_data_with_count['percentage'].round(2).astype(str) + '%)'

# Create the sunburst chart with the percentage in the labels
fig = px.sunburst(
    sunburst_data_with_count,
    path=['Final_Label', 'Higher_Topic_Label', 'Human_Readable_Topic'],  # Hierarchical path
    values='Count',  # Use count for size
    title="Three-Level Topic Hierarchy with Percentages",
    width=1200,  # Set width of the figure
    height=1200  # Set height of the figure
)

# Show the plot
fig.show()

```

```python
import plotly.express as px
import pandas as pd

# Load both CSV files
file_path_first = 'First_Step_BERTopic_topic_info_labelled_with_Final_Label_Merged.csv'
file_path_second = 'Second_Step_BERTopic_topic_info_labelled_with_Final_Label_Merged.csv'

data_first = pd.read_csv(file_path_first)
data_second = pd.read_csv(file_path_second)

# Clean the first dataset (remove duplicates and outliers)
data_first_cleaned = data_first.drop_duplicates(subset=['Name'])
data_first_cleaned = data_first_cleaned[data_first_cleaned['Topic'] != -1]

# Clean the second dataset (remove duplicates and outliers)
data_second_cleaned = data_second.drop_duplicates(subset=['Name'])
data_second_cleaned = data_second_cleaned[data_second_cleaned['Topic'] != -1]

# Concatenate both datasets
combined_data = pd.concat([data_first_cleaned, data_second_cleaned])

# Prepare data for sunburst chart
sunburst_data_combined = combined_data[['Final_Label', 'Higher_Topic_Label', 'Human_Readable_Topic', 'Count']].copy()

# Create the sunburst chart without percentages
fig = px.sunburst(
    sunburst_data_combined,
    path=['Final_Label', 'Higher_Topic_Label', 'Human_Readable_Topic'],  # Hierarchical path
    values='Count',  # Use count for size
    title="2024 January-August arXiv Papers: Sunburst Chart of Hierchical Topic Knowledge Interpretation",
    width=1000,  # Set width of the figure
    height=1000  # Set height of the figure
)

# Save the plot as an HTML file
html_file_combined = 'combined_sunburst_chart_without_percentages.html'
fig.write_html(html_file_combined)

# Output the file path for confirmation
html_file_combined
```

```python
import pandas as pd
import plotly.express as px

# Load both CSV files
file_path_first = 'First_Step_BERTopic_topic_info_labelled_with_Final_Label_Merged.csv'
file_path_second = 'Second_Step_BERTopic_topic_info_labelled_with_Final_Label_Merged.csv'

data_first = pd.read_csv(file_path_first)
data_second = pd.read_csv(file_path_second)

# Clean the first dataset (remove duplicates and outliers)
data_first_cleaned = data_first.drop_duplicates(subset=['Name'])
data_first_cleaned = data_first_cleaned[data_first_cleaned['Topic'] != -1]

# Clean the second dataset (remove duplicates and outliers)
data_second_cleaned = data_second.drop_duplicates(subset=['Name'])
data_second_cleaned = data_second_cleaned[data_second_cleaned['Topic'] != -1]

# Concatenate both datasets
combined_data = pd.concat([data_first_cleaned, data_second_cleaned])

# Prepare data for sunburst chart
sunburst_data_combined = combined_data[['Final_Label', 'Higher_Topic_Label', 'Human_Readable_Topic', 'Count']].copy()

# Create the sunburst chart without percentages
fig = px.sunburst(
    sunburst_data_combined,
    path=['Final_Label', 'Higher_Topic_Label', 'Human_Readable_Topic'],  # Hierarchical path
    values='Count',  # Use count for size
    title="2024 January-August arXiv Papers: Sunburst Chart of Hierchical Topic Knowledge Interpretation",
    width=1000,  # Set width of the figure
    height=1000  # Set height of the figure
)

# Save the plot as an HTML file
html_file_combined = 'combined_sunburst_chart_without_percentages.html'
fig.write_html(html_file_combined)

# Save the plot as a high-quality PNG file
# Increase scale for higher resolution (e.g., scale=5 will increase resolution by 5x)
png_file_combined = 'combined_sunburst_chart_without_percentages_high_quality.png'
fig.write_image(png_file_combined, scale=5)  # Higher scale gives better quality

# Output the file paths for confirmation
print(f"HTML file saved at: {html_file_combined}")
print(f"High-quality PNG file saved at: {png_file_combined}")
```
