# WIZMAP

## Overview
This script is designed for processing textual data related to Biden's context claims, encoding the text using a pre-trained embedding model, reducing the dimensionality for visualization, and finally visualizing the data using UMAP and an interactive web tool, Wizmap. 
The primary of this tool is to quickly analyze and visualize through embeddings Biden-related claims for better understanding and insight.

## Steps and Libraries Used

### Import Necessary Libraries
The script starts by importing required Python libraries for file handling, data processing, visualization, machine learning, web requests, and more.

- `glob`, `os.path`: For file path operations.
- `tqdm`: For progress bars.
- `json`: For loading and dumping JSON files.
- `matplotlib.pyplot as plt`: For plotting graphs.
- `collections.Counter`: For counting hashable objects.
- `sklearn.feature_extraction.text`: For converting text to vector form.
- `quadtreed3`: For spatial indexing and retrieval.
- `scipy.sparse`: For sparse matrix representation.
- `sklearn.neighbors`: For density estimation.
- `scipy.stats`: For statistical functions.
- `typing`: For type hints.
- `io`: For handling byte streams.
- `umap`: For dimensionality reduction.
- `pandas`, `numpy`: For data manipulation.
- `ndjson`, `requests`, `urllib`: For web requests and handling NDJSON files.
- `wizmap`: For generating interactive visualizations.
- `sentence_transformers`: For sentence embeddings.
- `datetime`: For handling date and time.

### Data Preparation
- A seed for reproducibility is set.
- DPI setting is adjusted for higher resolution plots. 
- A CSV file containing Biden-related claims is read into a Pandas DataFrame.

### Embeddings Generation
- A pre-trained sentence embedding model is loaded.
- The text data is encoded into embeddings in batches, with a progress bar showing the process.

### Dimensionality Reduction
- UMAP (Uniform Manifold Approximation and Projection) is used to reduce the embeddings to 2 dimensions for visualization purposes, using cosine metric.

### Data Visualization
- The 2D embeddings are visualized using a scatter plot, indicating Biden context-related claims.

### Interactive Visualization Preparation
- Coordinates (x, y) and texts from the reduced embeddings are prepared.
- These are used to generate a data list and a grid dictionary for interactive visualization using Wizmap.

### Saving and Displaying the Interactive Visualization
- The generated data list and grid dictionary are saved as JSON files.
- URLs for the NDJSON and JSON data are defined.
- Wizmap is used to visualize the data interactively on a web interface.

## Conclusion
This script effectively demonstrates the workflow from data loading, processing, embedding, dimensionality reduction, to interactive visualization for Biden-related claims. It utilizes advanced techniques and tools for a comprehensive analysis and presentation.


```python
# Import all the necessary libriaries
from glob import glob
from os.path import exists, join, basename
from tqdm import tqdm
from json import load, dump
from matplotlib import pyplot as plt
from collections import Counter

from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from quadtreed3 import Quadtree, Node
from scipy.sparse import csr_matrix
from sklearn.neighbors import KernelDensity
from scipy.stats import norm
from typing import Tuple
from io import BytesIO
from umap import UMAP

import pandas as pd
import numpy as np
import ndjson
import requests
import urllib
import wizmap

import datetime 


SEED = 20230501

plt.rcParams['figure.dpi'] = 300
```

```python
arxiv = pd.read_csv('filtered_arxiv_metadata_2024.csv')
```

```python
# arxiv = arxiv.sample(n=50000, random_state=42)
claims_en = arxiv.copy()
claims_en
```

```python
# Load the pre-trained embedding model
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

claims_en = claims_en.reset_index(drop=True)

# Encode all 25k reviews
BATCH_SIZE = 128
embeddings = model.encode(claims_en['abstract'], batch_size=BATCH_SIZE, show_progress_bar=True)
```

```python
fact_checks_title_and_claims_numpy = claims_en['title'].values
```

Dimensionality Reduction

```python
reducer = UMAP(metric='cosine')
embeddings_2d = reducer.fit_transform(embeddings)
```

```python
plt.title(f'UMAP Projected Embeddings of {len(fact_checks_title_and_claims_numpy)}Arxiv Related Data')
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], s=0.1, alpha=0.2)
plt.show()
```

## 3. Generate Two JSON Files for WizMap


```python
xs = embeddings_2d[:, 0].astype(float).tolist()
ys = embeddings_2d[:, 1].astype(float).tolist()
texts = fact_checks_title_and_claims_numpy
```

```python
data_list = wizmap.generate_data_list(xs, ys, texts)
grid_dict = wizmap.generate_grid_dict(xs, ys, texts, 'Arxiv Related Data')
```

```python
# Save the JSON files
wizmap.save_json_files(data_list, grid_dict, output_dir='./')
```

```python
data_url = 'https://thomyks.github.io/factcheck.github.io/arxiv_abstract_2024.ndjson'
grid_url = 'https://thomyks.github.io/factcheck.github.io/arxiv_abstract_2024.json'
```

```python
# Display wizmap
wizmap.visualize(data_url, grid_url, height=700)
```
