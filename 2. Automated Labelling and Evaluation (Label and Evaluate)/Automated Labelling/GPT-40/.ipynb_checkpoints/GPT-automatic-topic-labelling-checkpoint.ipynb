```python
import openai
import csv
import json
import os
import time  # Needed for sleeping

openai.api_key = ''

# Path to your CSV file (BERTopic results)
csv_file_path = 'First_Step_BERTopic_topic_info.csv'

# Updated CSV file path (Labelled BERTopic results)
updated_csv_file_path = 'First_Step_BERTopic_topic_info_labelled.csv'

# Read the CSV file
with open(csv_file_path, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    rows = list(reader)

# Process each row
for row in rows:
    while True:  # Keep trying until successful
        try:
            # Extract the necessary columns
            topic = row['Topic']
            representation = row['Representation']
            representative_docs = row['Representative_Docs']

            # Generate the human-readable label using OpenAI
            response = openai.ChatCompletion.create(
                model="gpt-4o",
                messages=[{
                    "role": "user",
                    "content": (f"You are an intelligent assistant skilled in generating labels for topics. "
                                f"Given the following details about a topic: "
                                f"Non-human readable Topic: {topic}. "
                                f"Representative Documents: {representative_docs}, "
                                f"Keywords: {representation}. "
                                f"Please generate a concise, human-readable, more general label for this topic trying to avoid countries. "
                                "Return only the label.")
                }]
            )

            # Extract the response
            label = response.get('choices', [{}])[0].get('message', {}).get('content', '').strip()

            # Update the row with the generated data
            row['Human_Readable_Topic'] = label
            print(label)

            break  # Exit the loop if successful

        except openai.error.RateLimitError as e:
            print(f"Rate limit reached, waiting to retry... {e}")
            time.sleep(60)  # Wait for 1 minute before retrying
        except Exception as e:
            print(f"An error occurred while processing row: {e}")
            break  # Exit the loop on other errors

# Prepare for writing to the new CSV by including the new 'Human_Readable_Topic' field
fieldnames = reader.fieldnames + ['Human_Readable_Topic']

# Write the updated data to a new CSV file
with open(updated_csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)
```
